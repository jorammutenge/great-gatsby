```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

import zipfile
import collections
import re
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
#View the data
df = pd.read_excel('greatGatsby.xlsx')
display(df.head())
print(f'Total no. of rows: {df.shape[0]}')
```

```python
#Choose relevant column
cols = ['title', 'rating', 'body']

#Assign sentiment to rating
def set_sentiment(df, rating_, sentiment_):
    df.loc[df['sentiment'] == rating_, 'sentiment'] = sentiment_
    return df

#Create clean df
df = (df
     [cols]
     .assign(body = df[['title', 'body']].agg(lambda x: '. '.join(x.values), axis=1).T
     )
     .rename(columns = {'body': 'review', 'rating': 'sentiment'})
     .drop(['title'], axis=1)
     .pipe(set_sentiment, 1, 'extremely negative')
     .pipe(set_sentiment, 2, 'negative')
     .pipe(set_sentiment, 3, 'neutral')
     .pipe(set_sentiment, 4, 'positive')
     .pipe(set_sentiment, 5, 'extremely positive')
     )
```

```python
#Check distribution of sentiment
df.sentiment.value_counts().plot(kind = 'barh');
```

```python
#Convert extremely positive to positive. Repeat for negative.
df.loc[df["sentiment"] == "extremely negative", "sentiment"] = "negative"
df.loc[df["sentiment"] == "extremely positive", "sentiment"] = "positive"
df = df.reset_index(drop=True)

# distribution of labels after collapsing labels.
labels = df["sentiment"].tolist()
labels_counter=collections.Counter(labels)
print("Label distribution: ",labels_counter)
```

```python
#Convert to lowercase and remove special characters.
def basic_transformation(df):    
    df['OriginalReview'] = df['review'].str.lower()
    df['CleanedReview'] = ""
    
    for index, row in df.iterrows():
        review = row["OriginalReview"]
        df.at[index,"CleanedReview"] = re.sub(r'[^A-Za-z0-9]+', ' ',str(review))
        
    return df

df = basic_transformation(df)
print(df.review.tolist()[19])
```

```python
#Tokenize the text.
from nltk.tokenize import word_tokenize  

def tokenization(df):    
    df['TokenizedReview'] = ""
    
    for index, row in df.iterrows():
        review = row["CleanedReview"]     
        # tokenize
        words = word_tokenize(review)
        df.at[index,"TokenizedReview"] = words                            
    return df
               
df = tokenization(df)
print(df.TokenizedReview.tolist()[19])
```

```python
from collections import Counter
def top_n_stemmed_word(n,df, column,label):
    stemmed_review = df[df["sentiment"] == label][column].tolist()    
    flat_list = [item for sublist in stemmed_review for item in sublist]
    most_common_words= [word for word, word_count in Counter(flat_list).most_common(n)]
    return most_common_words

top_15_Positive = top_n_stemmed_word(15,df,"TokenizedReview","positive")
top_15_Negative = top_n_stemmed_word(15,df,"TokenizedReview","negative")

print(f"Top 15 positive words:\n{top_15_Positive}\n")
print(f"Top 15 negative words:\n{top_15_Negative}")
```

```python
from nltk.corpus import stopwords  
stop_words = set(stopwords.words('english'))  

def remove_stopwords(df):    

    df['StopwordRemovedReview'] = ""
    
    for index, row in df.iterrows():
        words = row["TokenizedReview"]
        removed_stop_words = []
        for word in words:
            if word not in stop_words:
                removed_stop_words.append(word)        
        df.at[index,"StopwordRemovedReview"] = removed_stop_words 
    return df

df = remove_stopwords(df)
print(df.StopwordRemovedReview.tolist()[19])
```

```python
#Create a function for stemming 
from nltk.stem import PorterStemmer 
ps = PorterStemmer() 

def stemming(df):
    df['StemmedReview'] = ""
    df['ProcessedReview'] = ""
    
    #stemming
    for index, row in df.iterrows():
        stemmed_words = []
        words = row["TokenizedReview"] 
        altered_text = ""
        for word in words:
            stemmed_word = ps.stem(word)
            stemmed_words.append(stemmed_word)
            altered_text = altered_text + " " +stemmed_word
        df.at[index,"StemmedReview"] = stemmed_words
        df.at[index,"ProcessedReview"] = altered_text
    return df
df = stemming(df)
print(f'{df.StemmedReview.tolist()[19]}\n')
print(df.ProcessedReview.tolist()[19])
```
